We define the Ishigami function as: $$f(\mathbf{x})=sin(x_1)+a*sin^2(x_2)+b*x_3^4*sin(x_1)$$

where $$\mathbf{x}=\{x_1,x_2,x_3\} \in [-\pi,\pi]$$ and $$a = 5.0, b = 0.1$$ following https://mlatcl.github.io/mlphysical/lectures/05-01-sensitivity-analysis.html

```python
from scipy.stats.distributions import uniform
import numpy as np
from pyDOE import *
np.random.seed(42) #for reproducibility purposes

# Defining the Ishigami function
def ishigami(x,a=5,b=0.1):
    return np.sin(x[0])+a*np.sin(x[1])**2+b*np.sin(x[0])*x[2]**4

lower_bounds = [-3.14, -3.14, -3.14]
upper_bounds = [3.14, 3.14, 3.14]

n_samples = 10000
samples = lhs(len(lower_bounds), samples=n_samples)

#Latin Hypercube Sampling of x1,x2,x3
for i in range(len(lower_bounds)):
    samples[:, i] = uniform(loc=lower_bounds[i], scale=np.subtract(
        upper_bounds[i], lower_bounds[i])).ppf(samples[:, i])

results = []
for sample in samples:
    results.append(ishigami(sample))

print("Mininum y:", np.min(results))
print("Maximum y:", np.max(results))
print("Distance:", np.max(results)-np.min(results))
```

```python
import matplotlib.pyplot as plt

plt.scatter(samples[:,0],results)
plt.xlabel('x1')
plt.ylabel('y')
plt.savefig("baseline-x1.png",bbox_inches='tight')
plt.show()
plt.close()
plt.scatter(samples[:,1],results)
plt.xlabel('x2')
plt.ylabel('y')
plt.savefig("baseline-x2.png",bbox_inches='tight')
plt.show()
plt.close()
plt.scatter(samples[:, 2], results)
plt.xlabel('x3')
plt.ylabel('y')
plt.savefig("baseline-x3.png",bbox_inches='tight')
plt.show()
plt.close()
```
